```python
import numpy as np

x = np.zeros(20)
y = np.zeros(20)

y[-1] = 0
x[-1] = 0
x[0] = 1
x[1] = -1

for n in range(0,4):
    y[n] = -0.5*y[n-1]+x[n]+2*x[n-1]
    
print(y[0:4])
```

```python
from scipy import signal
x = np.array([1,-1,0,0],dtype=float)
b = [1,2]
a = [1,0.5]
y = signal.lfilter(b,a,x)
print(y)
```

```python
from matplotlib import pyplot as plt
n = np.arange(0,4)
plt.stem(n,y)
plt.title('y[n]')
plt.xlabel('n')
```

```python
import numpy as np 

x=np.array([1,2,-1],dtype=float) 
h=np.array([1,3],dtype=float) 
nXmin = 0 
nHmin = 0 

M = len(x) 
N = len(h) 
L = M + N -1 
y=np.zeros(L) 

for n in range(L):     
    y[n]=0 
    for k in range(M): 
        if (n-k)>=0 and (n-k)<N: 
            y[n]=y[n]+x[k]*h[n-k]
            
print(y)
```

```python
nY = [i for i in range(nXmin+nHmin,nXmin+nHmin+L)]
print(nY)
```

```python
h2 = np.zeros(26) # System-2 definied
h2[3] = 1
h2[4] = 1
print(h2)
```

```python
n = np.arange(26)
x3 = np.zeros(26)

for i in n:
    if (i>19):
        x3[i] = 0
    else:
        x3[i] = np.cos(0.2*np.pi*i)

plt.stem(n,x3)
plt.title('x3[n]')
```

```python
y = signal.convolve(x3,h2)
print(y)
```

```python
y = y[0:26]
plt.stem(n,y)
plt.title('y3[n]')
```
