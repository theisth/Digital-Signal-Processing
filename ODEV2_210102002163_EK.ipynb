# a)

```python
n = np.arange(10,20)

x = np.cos(0.2*np.pi*n)
h = np.array([1/3,1/3,1/3],dtype=float)

nXmin = 10
nHmin = -1

M = len(x)
N = len(h)
L = M + N - 1
y = np.zeros(L)

for i in range(L):
    y[i] = 0
    for k in range(M):
        if (i-k) >= 0 and (i-k) < N:
            y[i] = y[i] + x[k]*h[i-k]

print(y)
```

```python
nY = [i for i in range(nXmin+nHmin,nXmin+nHmin+L)]
print(nY)
```

```python
plt.stem(nY,y)
plt.title('y[n]')

plt.figure()

plt.stem(n,x)
plt.title('x[n]')
```
