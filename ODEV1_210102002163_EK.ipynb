# SORU - 1

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Aşağıdaki gibi verilerimiz tanımlanmıştır.
a = 3 - 4j
b = 1 + 2j
C = a + b
print(C) # Sonuç print ile yazıldı.
```

# I)

```python
# Kompleks sayinin real degeri np.real() ile bulunmustur. 
print(np.real(C))
```

# II)

```python
# Kompleks sayinin sanal kismi icin np.imag() ile bulunmustur.
print(np.imag(C))
```

# III)

```python
# Kompleks sayinin genligi icin np.abs() fonksiyonu kullanilmistir.
print(np.abs(C))
```

# IV)

```python
# Kompleks sayinin faz açısı np.angle() ile bulunur.
print(np.angle(C))
```
