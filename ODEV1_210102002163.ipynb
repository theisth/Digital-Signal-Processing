```python

# 1. sinus dalgası == > 0.4*pi*N = 2*pi*k ==> N1 = 5s periyotlu
# 2. sinus dalgası == > 0.6*pi.N = 2*pi*k ==> N2 = 10s periyotlu

# Bu iki sünus dalgasından x[n] elde edildiğinden x[n] periyodu N1 ve N2'nin
# EKOK'larıdır.
#                        Tx = EKOK(5,10) = 10s 'dir

# Grafiğimiz iki periyot olduğundan 20 saniyelik bir grafik çizdirilmelidir.
```

```python
n = np.arange(0,20) # 0'dan 20'ye tanım aralığı kullanılmıştır.
x = 2*np.sin(0.4*np.pi*n - 0.2*np.pi) + np.sin(0.6*np.pi*n - 0.4*np.pi) #x[n]
plt.stem(n,x)
plt.show()
```

```python
# Yukarıdaki tabloda görüldüğü üzere 2 periyot görülmektedir.
# 0'dan 10'a olan kısım ile 10'dan 20'e olan kısım boyunca simetriktir.
# Bu yüzden sistemin periyodu 10 saniyedir.
```

# I)

```python
x2 = np.exp(1j*np.pi*0.2)
n2 = np.arange(0,16)
print(x2)
```

```python
# Yukarıda görüldüğü üzere x2[n] işaretinin sanal ve real kısımları mevcuttur.
# Real ve sanal kısımları ayrı ayrı çizdirilmiştir.

x2_image = np.zeros(16) # 0'dan 15'e kadar bir 0'lardan oluşan array açıldı. 
for i in range(0,16):
    x2_image[i] = np.imag(x2)
plt.title('x[2] sanal değerleri')
plt.xlabel('n')
plt.ylabel('x2[n] sanal değerleri')
plt.stem(n2,x2_image)
plt.show()
```

```python
x2_real = np.zeros(16)
for i in range(0,16):
    x2_real[i] = np.real(x2)
plt.title('x[2] real değerleri')
plt.xlabel('n')
plt.ylabel('x2[n] real değerleri')
plt.stem(n2,x2_real)
plt.show()
```
